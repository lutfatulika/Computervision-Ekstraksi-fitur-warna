# **3.Impoert Libary**

```python
import cv2              # untuk olah citra
import os               # untuk akses folder
import numpy as np      # untuk perhitungan numerik
import pandas as pd     # untuk dataframe & simpan file
```

# **4. Menentukan Path Dataset**

```python
# Path folder dataset di Google Drive
folder_path = '/content/drive/MyDrive/SEMESTER 6/COMVIS/TUGAS 2'
```

# 4.Prepocessing & Ektraksi Fitur warna RGB

```python
data = []

# Loop semua gambar dalam folder
for filename in os.listdir(folder_path):

    # Gabungkan path
    img_path = os.path.join(folder_path, filename)

    # Baca gambar
    img = cv2.imread(img_path)

    # Jika bukan gambar, skip
    if img is None:
        continue

    # Resize agar semua gambar seragam
    img = cv2.resize(img, (256, 256))

    # Hitung rata-rata warna (RGB)
    mean_b = np.mean(img[:, :, 0])  # Blue
    mean_g = np.mean(img[:, :, 1])  # Green
    mean_r = np.mean(img[:, :, 2])  # Red

    # Simpan ke list
    data.append([filename, mean_r, mean_g, mean_b])

# Buat DataFrame
df = pd.DataFrame(data, columns=["Image", "R", "G", "B"])

# Tampilkan hasil
print(df)
```

# **5.Menampilkan jumlah nilai RGB**

```python
# Menjumlahkan seluruh nilai RGB
total_rgb = df[["R", "G", "B"]].sum()

print("Total Nilai RGB:")
print(total_rgb)
```

## **6.Menyimpan Ke Excel**

```python
# Simpan ke Excel
excel_path = folder_path + "/fitur_warna.xlsx"
df.to_excel(excel_path, index=False)

print("File berhasil disimpan!")
```
